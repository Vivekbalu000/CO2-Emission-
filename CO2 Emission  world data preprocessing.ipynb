# 1021 Null values was replaced by fillna

```python
dataset[' Gdp'].fillna(0,inplace=True)
```

# 301 Null values was replaced by fillna

```python
dataset['Lifeexpectation'].fillna(dataset['Lifeexpectation'].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset[' Yearlychange'].replace(to_replace=['Err:520'],value='0',inplace=True)
```

```python
dataset[' Population'] = dataset[' Population'].str.replace (',','', regex=True)
```

# Data type changed from object to float

```python
dataset[' Population'] = dataset[' Population'].astype('float64')
```

```python
dataset[' Yearlychange'] = dataset[' Yearlychange'].astype('float64')
```
